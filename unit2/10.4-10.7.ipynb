```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('tips.csv')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_title("Общая сумма счета")
axes.set_xlabel('Сумма заказа, USD')
axes.set_ylabel('Количество заказов')
axes.hist(df['total_bill'],bins=20,color='green')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,1])
axes.scatter(x=df['total_bill'],y=df['tip'])
axes.set_xlabel('сумма заказа')
axes.set_ylabel('сумма чаевых')
```

```python
fig = plt.figure()
main_axes = fig.add_axes([0,0,1,1])
additional_axes = fig.add_axes([0.6,0.6,0.38,0.38])
main_axes.hist(df['total_bill'])
min_max_day_total_bill = df.groupby('day')['total_bill'].agg(['min','max']).reset_index()
additional_axes.bar(x=min_max_day_total_bill['day'],height=min_max_day_total_bill['max'],width = 0.4, align = 'edge')
additional_axes.bar(x=min_max_day_total_bill['day'],height=min_max_day_total_bill['min'],width = -0.4, align = 'edge')
```

```python
fig = plt.figure()
main_axes = fig.add_axes([0,0,1,1])
min_max_day_total_bill = df.groupby('day')['total_bill'].agg(['min','max']).reset_index()
main_axes.bar(x=min_max_day_total_bill['day'],height=min_max_day_total_bill['max'],width = 0.4, align = 'edge', label="Максимальная сумма чека")
main_axes.bar(x=min_max_day_total_bill['day'],height=min_max_day_total_bill['min'],width = -0.4, align = 'edge', label="Минимальная сумма чека")
axes.legend(loc=3)
```
