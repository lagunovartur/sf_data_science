```python
import json
import pandas as pd
from pprint import pprint
```

```python
with open('recipes.json') as f:
    recipes = json.load(f)
```

```python
pprint(recipes)
```

```python
food = {}
for recipe in recipes:
    for ingredient in recipe['ingredients']:
        if ingredient in food.keys():
            food[ingredient] += 1
        else:
            food[ingredient] = 1
```

```python
pprint(food)
```

```python
for item in food:
  if food[item] > 100:
      print(item)


```

```python
max = ""
maxValue = 0

for item in food:
   if food[item] >  maxValue:
       max = item
       maxValue = food[item]

print(max)
```

```python
df = pd.DataFrame(recipes)
df.head()
```

```python
df2 = df.explode('ingredients')
df2.head()
```

```python
df3 = pd.get_dummies(data=df2, columns=['ingredients'], prefix='',prefix_sep='')
df3.head()
```

```python
df4 = df.copy()

def find_item(cell):
    return int(item in cell)

for item in food:
    df4[item] = df4['ingredients'].apply(lambda x: find_item(x))

df4['ingredients'] = df4['ingredients'].apply(lambda x: len(x))

print(df4.head())
```

```python
df4.to_csv('recipes2.csv',index=False)
```
